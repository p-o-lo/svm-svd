```julia
using HDF5
using LinearAlgebra
K = h5read("NLO_3B_ex_with3B_12_12.h5", "kinetic")
V₂ = h5read("NLO_3B_ex_with3B_12_12.h5", "twoBodyPotential")
V₃ = h5read("NLO_3B_ex_with3B_12_12.h5", "threeBodyPotential")
S = h5read("NLO_3B_ex_with3B_12_12.h5", "overlaps")
alphas = h5read("NLO_3B_ex_with3B_12_12.h5", "alphas")
H = K + V₂ + V₃
E = eigen(H,S)
sortE = sort(E.values)
indexOfground = argmin(E.values)
indexOfFirstEx = findall(x->x==sortE[2],E.values)
C0 = E.vectors[:,indexOfground]
C1 = E.vectors[:,indexOfFirstEx]
function ψ₀(r₁,r₂,r₃)
    wf = 0
    for i = 1:size(E.values,1)
        wf += C0[i]*exp(-0.5*(alphas[1,i]*dot(r₁-r₂,r₁-r₂) + alphas[2,i]*dot(r₁ - r₃,r₁ - r₃) + alphas[3,i]*dot(r₂-r₃,r₂-r₃)))
    end
    return abs(wf)
end
function ψ₁(r₁,r₂,r₃)
    wf = 0
    for i = 1:size(E.values,1)
        wf += C1[i]*exp(-0.5*(alphas[1,i]*dot(r₁-r₂,r₁-r₂) + alphas[2,i]*dot(r₁ - r₃,r₁ - r₃) + alphas[3,i]*dot(r₂-r₃,r₂-r₃)))
    end
    return wf
end
# x₁ = -60 + 120*rand()
# y₁ = -60 + 120*rand()
# r₁ = [x₁, y₁]
# x₂ = -100 + 200*rand()
# y₂ = -100 + 200*rand()
# r₂ = [x₂, y₂]
# x₃ = -(x₁ + x₂)
# y₃ = -(y₁ + y₂)
# r₃ = [x₃, y₃]
# r₁-r₂
# ψ₁(r₁,r₂,r₃)
```

```julia
using Plots
pyplot()
using LinearAlgebra
X = zeros((3,2000))
Y = zeros((3,2000))
Xr = zeros((3,2000))
Yr = zeros((3,2000))
Xp = zeros((3,2000))
Yp = zeros((3,2000))
I = zeros((2,2))
xs = range(-70,70,length = 10)
axesPlt = plot()
pointPlt = plot()
vline!([0,0], line=(:black))
hline!([0,0], line=(:black))
rplt = plot()
vline!([0,0], line=(:black))
hline!([0,0], line=(:black))
prplt = plot()
vline!([0,0], line=(:black))
hline!([0,0], line=(:black))
for i = 1:2000
    x₁ = -1. + 2*rand()
    y₁ = -1. + 2*rand()
    r₁ = [x₁, y₁]
    x₂ = -1. + 2*rand()
    y₂ = -1. + 2*rand()
    r₂ = [x₂, y₂]
    x₃ = -(x₁ + x₂)
    y₃ = -(y₁ + y₂)
    r₃ = [x₃, y₃]
    prob = ψ₁(r₁,r₂,r₃)^2
    #print(prob,"\n")
    X[:,i] = [x₁,x₂,x₃]
    Y[:,i] = [y₁,y₂,y₃]
    pointPlt = scatter!(pointPlt, X[:,i],Y[:,i], leg=false)
    I[1,1] = x₁^2 + x₂^2 + x₃^2
    I[2,2] = y₁^2 + y₂^2 + y₃^2
    I[1,2] = -x₁*y₁ - x₂*y₂ - x₃*y₃
    I[2,1] = -x₁*y₁ - x₂*y₂ - x₃*y₃
    Ip = eigen(I)
    indexLowInert = argmin(Ip.values)
    #print(Ip.vectors[:,indexLowInert])
    #axesPlt = plot!(pointPlt, xs, Ip.vectors[2,indexLowInert]/Ip.vectors[1,indexLowInert]*xs, leg=false)
    θ = atan(Ip.vectors[2,indexLowInert]/Ip.vectors[1,indexLowInert])
    #print(θ)
    R = [cos(3*pi/2 - θ) -sin(3*pi/2 - θ);sin(3*pi/2 - θ) cos(3*pi/2 - θ)]
    #print("\n",R)
    R₁ = R*r₁
    R₂ = R*r₂
    R₃ = R*r₃
    Xr[:,i] = [R₁[1],R₂[1],R₃[1]]
    Yr[:,i] = [R₁[2],R₂[2],R₃[2]]
    rplt = scatter!(rplt,Xr[:,i],Yr[:,i], leg=false)
    #print(R₁,R₂,R₃)
    while ((R₁[1] < 0. && R₁[2] > 0.) || (R₂[1] < 0. && R₂[2] > 0.) || (R₃[1] < 0. && R₃[2] > 0.))
        R₁[1] = -R₁[1]
        R₂[1] = -R₂[1]
        R₃[1] = -R₃[1]
        if ((R₁[1] < 0. && R₁[2] > 0.) || (R₂[1] < 0. && R₂[2] > 0.) || (R₃[1] < 0. && R₃[2] > 0.))
            R₁[2] = -R₁[2]
            R₂[2] = -R₂[2]
            R₃[2] = -R₃[2]
        end
    end
    Xp[:,i] = [R₁[1],R₂[1],R₃[1]]
    Yp[:,i] = [R₁[2],R₂[2],R₃[2]]
    #if (prob >= 0.001)
    prplt = scatter!(prplt,Xp[:,i],Yp[:,i], leg=false, markersize = 5, color=RGB(abs(prob-1.0e-9)/(1.0e-9),abs(prob-1.0e-9)/(1.0e-9),abs(prob-1.0e-9)/(1.0e-9)))
    #end
end
#display(axesPlt)
display(rplt)
display(prplt)
```
